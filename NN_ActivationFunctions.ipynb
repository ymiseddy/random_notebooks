## Softplus

```python
def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
fnplot(softplus, -5, 50)
```

## Sigmoid

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
fnplot(sigmoid, -5, 5)
```
