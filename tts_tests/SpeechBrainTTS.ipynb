```python
import torchaudio
from speechbrain.pretrained import Tacotron2
from speechbrain.pretrained import HIFIGAN
import IPython
```

```python
# Intialize TTS (tacotron2) and Vocoder (HiFIGAN)
#tacotron2 = Tacotron2.from_hparams(source="awacke1/Text-to-Speech-speechbrain-tts-tacotron2-ljspeech", savedir="tmpdir_tts"

tacotron2 = Tacotron2.from_hparams(source="speechbrain/tts-tacotron2-ljspeech", savedir="tmpdir_tts")
hifi_gan = HIFIGAN.from_hparams(source="microsoft/speecht5_hifigan", savedir="tmpdir_vocoder")
#hifi_gan = HIFIGAN.from_hparams(source="speechbrain/tts-hifigan-libritts-16kHz", savedir="tmpdir_vocoder")

#hifi_gan = HIFIGAN.from_hparams(source="speechbrain/tts-hifigan-ljspeech", savedir="tmpdir_vocoder")
```

```python
# Running the TTS
mel_output, mel_length, alignment = tacotron2.encode_text("Too many people, with too many problems and not enough love to go around.")

# Running Vocoder (spectrogram-to-waveform)
waveforms = hifi_gan.decode_batch(mel_output)
waveforms
```

```python
# Save the waverform
torchaudio.save('example_TTS_microsoft.wav',waveforms.squeeze(1), 22050)
#torchaudio.save('libbrits.wav',waveforms.squeeze(1), 22050)
```
